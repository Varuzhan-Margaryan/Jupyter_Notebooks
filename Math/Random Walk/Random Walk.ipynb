# Random Walk visualization

## 1-Dimentional case

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=(10,10)
plt.rcParams['lines.linewidth'] = 0.6
```

```python
def get_new_pos(last_pos):
    if np.random.rand()<0.5:
        return last_pos-1
    return last_pos+1
```

```python
last_pos = 0
plt.ylim(-50,50)
t=[0]
y=[0]
def animate(i):
    global last_pos
    last_pos = get_new_pos(last_pos)
    t.append(i)
    y.append(last_pos)
    plt.cla()
    plt.plot(t,y)
    
anim = FuncAnimation(plt.gcf(),animate,interval = 100)
```

## 2-Dimentional case

```python
def get_new_pos2(last_x,last_y):
    last_x+=2*np.random.rand()-1
    last_y+=2*np.random.rand()-1

    return last_x, last_y
```

```python
last_x = last_y = 0
m = 10
x = [last_x]
y = [last_y]
def animate(i):
    global last_x, last_y, m
    last_x, last_y = get_new_pos2(last_x, last_y)
    x.append(last_x)
    y.append(last_y)
    plt.cla()
    plt.title(f'distance = {np.round((last_x**2+last_y**2)**(1/2))}; iterations = {i}')
    plt.plot(x,y)
    plt.scatter(0,0, c='g')
    plt.scatter(last_x, last_y, c='b')
    m = max(m-1, abs(last_x), abs(last_y))+1
    plt.xlim(-m,m)
    plt.ylim(-m,m)

    
anim = FuncAnimation(plt.gcf(),animate,interval = 10)
```

## 3-Dimentional case

```python
def get_new_pos3(last_x,last_y,last_z):
    last_x+=2*np.random.rand()-1
    last_y+=2*np.random.rand()-1
    last_z+=2*np.random.rand()-1

    return last_x, last_y, last_z
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection = '3d')
last_x = last_y = last_z = 0
m = 10
x = [last_x]
y = [last_y]
z = [last_z]

def animate(i):
    global last_x, last_y, last_z, m
    last_x, last_y, last_z = get_new_pos3(last_x, last_y, last_z)
    x.append(last_x)
    y.append(last_y)
    z.append(last_z)
    plt.cla()
    plt.title(f'distance = {np.round(( last_x**2+last_y**2+last_z**2 )**(1/2))}; iterations = {i}')
    ax.scatter3D(0,0,0, c= 'g')
    ax.scatter3D(last_x,last_y,last_z , c= 'b')
    ax.plot3D(x,y,z)
    
    m = max(m-1, abs(last_x), abs(last_y), abs(last_z))+1
    ax.set_xlim(-m,m)
    ax.set_ylim(-m,m)
    ax.set_zlim(-m,m)
    
anim = FuncAnimation(fig,animate,interval = 10)
```
