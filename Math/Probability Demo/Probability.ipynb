```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
plt.style.use('ggplot')
plt.rcParams['figure.figsize']  = (10,5)
```

```python
#erku hogi kopek en qcum. um mot gir exav haxtum a. tesakanoren arajini haxtelu havanakanutyuny 2/3 e
n1 = 0
n2 = 0
turn = 0
games = 1
def animate(i):
    global n1,n2,turn,games
    turn+=1
    if np.random.random()>0.5:
        if turn%2==1:
            n1 += 1
        else:
            n2 += 1
        turn = 0
        games += 1

    y1 = n1/games
    y2 = n2/games
    plt.cla()
    plt.title(f'games = {games}')
    plt.xticks([1,2])
    plt.ylim(0,1)
    plt.xlim(1,2)
    plt.bar([1,2],[y1,y2])
    plt.plot([0, 3],[1/3, 1/3],c='b')
    plt.plot([0, 3],[2/3, 2/3],c='b')


    
animate(1)
animate(2)

anim = FuncAnimation(plt.gcf(),animate,interval = 1)
```
