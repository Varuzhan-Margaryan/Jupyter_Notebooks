# Math. Calculus

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.misc import derivative as dr
```

## Taylor series

```python
def taylor(n,f,x0):
    adds = []
    for i in range(n+1):
        adds.append(taylor_add(i, f, x0))
        
    def func(x):
        nonlocal adds
        s = 0
        for i in range(len(adds)):
            s += adds[i](x)
        return s
    return func
        
    
    
def taylor_add(k,f,x0):
    fk = nder(f,k)(x0)
    k_fac = math.factorial(k)
    return lambda x : (fk/k_fac) * (x-x0)**k
    
    
def der(f):
    return lambda x: dr(f,x)


def nder(f,n):
    ders = [f]
    for i in range(n):
        ders.append(der(ders[i]))
    return ders[n]
```

```python
x0 = 0
n = 4
f = lambda x: np.exp(x)

xs = np.linspace(-2,2,100)

for n in range(10):
    plt.plot(xs,f(xs), label = 'f(x)')
    g = taylor(n,f,x0)
    plt.plot(xs,g(xs), label = f'taylor {n}th series ')
    plt.ylim(0,10)



    plt.legend()
    plt.show()
```
