# Let's try to draw a graph with matplotlib library

```python
"""
Matplotlib demo
"""
import numpy as np
import matplotlib.pyplot as mpl

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
col = np.random.rand(n)
area = np.pi * (np.random.rand(n) * 15) ** 2

mpl.scatter(x, y, s=area, c=col, alpha=.5)
mpl.show()
```
