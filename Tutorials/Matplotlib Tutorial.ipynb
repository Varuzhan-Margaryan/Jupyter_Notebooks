```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
plt.plot([1,2,3],[4,5,6])
```

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x, y)
plt.show()
```

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x, y)
plt.title('Our first graph!')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

### Use fontdict to change font

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x, y)
plt.title('Our first graph!', fontdict={'fontname':'Roboto','fontsize':50})
plt.xlabel('X Axis', fontdict={'fontsize':20})
plt.ylabel('Y Axis')

plt.show()
```

### Use xticks and yticks for resizing graph

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x, y)
plt.xticks([1,2,5,6])
plt.yticks([1,2,4,5.5,6])


plt.show()
```

### Use label parameter and plt.legend() function for labeling graph

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y,label='2x')
plt.legend()
plt.show()
```

### Multyple graphs

```python
x = [1,2,3,4]
y1 = [1,2,3,4]
y2 = [2,4,9,16]
y3 = [1,8,27,64]
plt.plot(x, y1,x,y2,x,y3)
plt.show()
```

### Color parameter and 3rd parameter in plot() for colors

```python
x = [1,2,3]
y1 = [2,4,6]
y2 = [4,8,12]
y3 = [20,40,60]

plt.plot(x, y1,color='red')
plt.plot(x, y2,color='r')
plt.plot(x, y3,'r')

plt.show()
```

### Other parameters for customizing graph

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y,label='2x',color='green',linestyle='--',linewidth=3,marker='.',markersize=20,markeredgecolor='blue')
plt.legend()
plt.show()
```

```python
#shorthand notation
##[color][marker][line]
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y,'b^-.',label='2x')
#color = blue
#marker = ^
#linestyle = --
plt.legend()
plt.show()
```

## Using Numpy

```python
np.arange(0,10.5,0.5)
```

```python
x=np.arange(0,10.5,0.5)
plt.plot(x,x**2,'b+:',label='x^2')
plt.legend()
plt.show()
```

```python
x=np.arange(0,10.5,0.5)
plt.plot(x[:15],x[:15]**2,'b-',label='x^2')
plt.plot(x[14:],x[14:]**2,'b--')

plt.legend()
plt.show()
```

## Resizing graph

```python
x=np.arange(0,10.5,0.5)

plt.figure(figsize=(5,2),dpi=100)##dip is px per inch

plt.plot(x,x**2,label='x^2')
plt.legend()
plt.show()
```

## Saving graphs

```python
x=np.arange(0,10.5,0.5)


plt.plot(x,x**2,label='x^2')
plt.legend()
plt.savefig('graphs\mygraph.png',dip=300)
```

### Line Graph

We will print graph of gas prices of differenc countries. (data in 'data/gas_prices.csv')

```python
gas = pd.read_csv('data/gas_prices.csv')
gas
```

### For 3 countries

```python
years = gas['Year']

UK = gas['UK']
USA = gas['USA']
Canada = gas['Canada']

plt.figure(figsize=(10,5),dpi=100)

plt.title('Gas Prices over Time (in US Dollars)',fontdict={'fontsize':20})
plt.ylabel('Price')
plt.xlabel('Year')

plt.xticks(years[::3])

plt.plot(years, UK , 'b.-',label = 'UK')
plt.plot(years, USA , 'r.-', label = 'USA')
plt.plot(years, Canada , 'g.-', label = 'Canada')

plt.legend()
plt.show()
```

### For All

```python
years = gas['Year']

UK = gas['UK']
USA = gas['USA']
Canada = gas['Canada']

plt.figure(figsize=(10,7),dpi=200)
plt.title('Gas Prices over Time (in US Dollars)',fontdict={'fontname':'Roboto','fontsize':20})
plt.ylabel('Price')
plt.xlabel('Year')
plt.xticks(years[::3])

for country in gas:
    if country != 'Year':
        plt.plot(years, gas[country] ,marker = '.',label = country)

plt.legend()
plt.savefig('graphs/gas_price.png',dpi=300)

plt.show()
```

#### Weights

```python
fifa['Weight']
```

```python
[x.strip('lbs') if type(x) is str else x for x in fifa['Weight']]
```

```python
weights = [int(x.strip('lbs')) if type(x) is str else x for x in fifa['Weight']]
weights[:5]
```

```python
wei = np.array(weights)

wei[:5]
```

```python
(wei>150).sum()
```

```python
light = (wei<125).sum()
light_medium = ((wei>125)&(wei<150)).sum()
medium = ((wei>150)&(wei<175)).sum()
medium_heavy = ((wei>175)&(wei<200)).sum()
heavy = (wei>200).sum()
all_w = [light,light_medium,medium,medium_heavy,heavy]

labels = ['-125','125-150','150-175','175-200','200-']
explode = [0.4,0.1,0,0,0.4]
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=100)

plt.title('Weights of FIFA players')
plt.pie(all_w, labels = labels, autopct = '%.2f %%',pctdistance=0.8,explode = explode)
plt.show()
```
